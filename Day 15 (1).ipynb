```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

p = 0.5  
x = [0,1] 

pmf = bernoulli.pmf(p,x)

plt.scatter(x,pmf,color="skyblue")
plt.title("Bernouli distribution")
plt.xlabel("Bernoulli 's pmf value")
plt.ylabel("The possible outcomes")
plt.show()


```

```python
from scipy.stats import uniform 
a = 2
b = 5
x=np.linspace(a,b,200)
pdf = uniform.pdf(x,loc=a,scale=b-a)
print(pdf)
plt.bar(x,pdf,color="yellow")
plt.show()
```

```python
#exponential - average rate(lambda.e^-lambda)
from scipy.stats import expon
lambda_value = 2
x=np.linspace(0,10,50)
pdf = expon.pdf(x,scale=1/lambda_value)
print(pdf)
plt.bar(x,pdf,color="Red")
plt.show()
```

```python
def mean_variance(x,prob_x):
    E = sum(v * p  for v, p in zip(x,prob_x)) 
    E2 = sum((v**2) * p for v,p in zip(x,prob_x))
    var__value = E2 - E**2
    return E , var__value

x = [0,1,2]
prob_x = [0.3,0.5,0.2]
E , V = mean_variance(x,prob_x)
print(f"The mean value is {E} and the variance is {V} ")
```

```python
#central limit theorem
import numpy as np
import matplotlib.pyplot as plt

n_trials = 1000
n_dice = 30
sample_means = []

for i in range(n_trials):
    rolls = np.random.randint(1,7,n_dice)
    sample_means.append(np.mean(rolls))

plt.hist(sample_means,bins=30,color='skyblue',edgecolor='black',density=True)
plt.title('central Limit Theorem with Dice Rolls')
plt.xlabel('Sample Mean of 30 Rolls')
plt.ylabel('Density')
plt.grid(True)
plt.axvline(x=3.5, color='red', linestyle='--', label='Expected Mean(3.5)')
plt.legend()
plt.show()






```
