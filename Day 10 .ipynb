```python
# Day 10

#Dunder Methods - It is a special method that starts and ends with double underscores (__init__, __str__, __repr__, etc.).
# like
# __init__() - It is a constructor method that initializes the object when it is created.
# __str__() - It is used to return a string representation of the object.   
# __repr__() - It is used to return a string representation of the object that can be used to recreate the object.
# __add__() - It is used to add two objects together.
# __sub__() - It is used to subtract two objects.
# __mul__() - It is used to multiply two objects.
# __truediv__() - It is used to divide two objects.
# __floordiv__() - It is used to perform floor division on two objects.
# __mod__() - It is used to perform modulus operation on two objects.
# __pow__() - It is used to perform power operation on two objects.
# __len__() - It is used to return the length of the object.
# __getitem__() - It is used to get the item at the specified index.
# __setitem__() - It is used to set the item at the specified index.
# __delitem__() - It is used to delete the item at the specified index.
# __contains__() - It is used to check if the object contains the specified item.
# __iter__() - It is used to return an iterator object.
# __next__() - It is used to return the next item from the iterator object.
# __call__() - It is used to make the object callable like a function.
# __eq__() - It is used to check if two objects are equal.
# __ne__() - It is used to check if two objects are not equal.
# __lt__() - It is used to check if one object is less than another.
# __le__() - It is used to check if one object is less than or equal to another.
# __gt__() - It is used to check if one object is greater than another.
# __ge__() - It is used to check if one object is greater than or equal to another.
# __hash__() - It is used to return the hash value of the object.
# __bool__() - It is used to return the boolean value of the object.
# __del__() - It is used to delete the object.
# __enter__() - It is used to enter the context manager.
# __exit__() - It is used to exit the context manager.p
```

```python
from datetime import date    # It is used to grt the date from the system. it is a pack

class person():

    def __init__(self, name): #__init__ is used for 
        self.name = name

    def __str__(self): # __str__ only for string representation not integer
        return self.name         #it is used to print class object in a readable format
        
    def __repr__(self): # __repr__ is used for debugging and development
        return self.name          # it will print the memory location    
    
    def __del__(self): #__del__ is used for deleting the object
        return self.name        # it will print the message when the object is deleted

    def __add__(self): #__add__ is used for adding two objects
        return self.name        # it will print the message when the object is added
    
    def __mul__(self): # __mul__ is used for multiplying two objects
        return self.name
    
    def __sub__(self): # __sub__ is used for substracting two objects
        return self.name
    
    def __div__(self): # __div__ is used for divide two Objects
        return self.name
    
    def __eq__(self): # __eq__ is used for check the equality of Two objects
        return self.name
    
    def __lt__(self): # __lt__ is used for check the Less than condition
        return self.name
    
    def __gt__(self): # __gt__ is used for check the Greater than condition
        return self.name
    
    def __le__(self): # __le__ is used for check the Less than or Equal Too Condition
        return self.name
    
    def __ge__(self): # __ge__ is used for check the greater than or Equal Too Condition
        return self.name
    
    def __ne__(self): # __ne__ is used for check the Not Equal Condition
        return self.name       
    
    def __len__(self): # __len__ is used to get the length of the object
        return self.name        # it will return the length of the object

    def __abs__ (self): # __abs__ is used to get the absolute value of the object
        return self.name         # it will return the absolute value of the object

    def __contains__ (self): # __contains__ is used to check if the object contains a value
        return self.name          # it will return True if the object contains the value else False

    def __getitem__ (self): # __getitem__ is used to get the item from the object
        return self.name         # it will return the item from the object

    def __hash__ (self): # __hash__ is used to get the hash value of the object
        return self.name         # it will return the hash value of the object

    def __setitem__ (self): # __setitem__ is used to set the item in the object
        return self.name         # it will set the item in the object
    
    def __delitem__ (self): # __delitem__ is used to delete the item from the object
        return self.name         # it will delete the item from the object
    
    def __call__(self): # __call__ is used to call the object like a function
        return self.name         # it will call the object like a function
    
    def __iter__(self): # __iter__ is used to iterate the object
        return self.name         # it will return the iterator of the object
    
    def __next__(self): # __next__ is used to get the next item from the object
        return self.name           # it will return the next item from the object
    
    def __bool__(self): # __bool__ is used to check if the object is True or False
        return self.name     # it will return True if the object is not empty else
```
